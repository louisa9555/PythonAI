###  อ่าน API key

```python
import yaml

with open("env.yaml") as f:
    config = yaml.safe_load(f)
api_key = config["GEMINI_API_KEY"]
```

###  คุยแบบ text

```python
from google import genai

client = genai.Client(api_key=api_key)

response = client.models.generate_content(
    model="gemini-2.5-pro",
    contents="Do you know LouisMakerLab youtube channel",
)

print(response.text)
```

### คุยแบบ text และ output รูปภาพ

```python
from google import genai
from google.genai import types
from PIL import Image
from io import BytesIO

client = genai.Client(api_key=api_key)

prompt = (
    "Create me a cat holding frappe"
)

response = client.models.generate_content(
    model="gemini-2.5-flash-image-preview",
    contents=[prompt],
)

for part in response.candidates[0].content.parts:
    if part.text is not None:
        print(part.text)
    elif part.inline_data is not None:
        image = Image.open(BytesIO(part.inline_data.data))
        image.save("./results/frappe_cat.png")
```

### Input รูปภาพ + text

```python
from google import genai
from google.genai import types
from PIL import Image
from io import BytesIO


client = genai.Client(api_key=api_key)
image_input = Image.open('./data/louis.jpg')
text_input = """ใส่ prompt ตรงนี้"""

# Generate an image from a text prompt
response = client.models.generate_content(
    model="gemini-2.5-flash-image-preview",
    contents=[text_input, image_input],
)


image_parts = [part.inline_data.data for part in response.candidates[0].content.parts if part.inline_data]
if image_parts:
    image = Image.open(BytesIO(image_parts[0]))
    image.save('./results/image.png')
    image.show()
```
